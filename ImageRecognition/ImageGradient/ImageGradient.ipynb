## Simple Example for Image Gradient

![TargetPixel](TargetPixel.png)

It is an example for understanding image gradient. 
The target pixel and surrounding pixels are listed.


```python
import numpy as np
from scipy import signal as sig
```

```python
data = np.array([[0, 70, 0],
                 [180, 255, 90],
                 [0, 40, 0]])
```

```python
G_x = sig.convolve2d(data, np.array([[0, 0, 0],
                                     [1, 0, -1],
                                     [0, 0, 0]]), mode='same')

G_y = sig.convolve2d(data, np.array([[0, -1, 0],
                                     [0, 0, 0],
                                     [0, 1, 0]]), mode='same')
```

```python
print(f'G_x is \n{G_x}')
print(f'G_y is \n{G_y}')
```

## Using Sobel Operator by opencv

[Sobel operator](https://en.wikipedia.org/wiki/Sobel_operator) is a common operator for image processing. It emphasizes the impact of directly adjacent pixels, which get assigned with higher weights.

We don't need to define kernels by ourselves. Using the **Sobel Operator** by opencv.

Here is a kitties picture.

![kitties](kitties.jpg)

```python
import numpy as np
import cv2
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("kitties.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Define the Sobel operator kernels.

kernel_x = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

kernel_y = np.array([[-1, -2, -1],
                     [0, 0, 0],
                     [1, 2, 1]])
```

```python
sobelG_x = sig.convolve2d(img, kernel_x, mode='same')
sobelG_y = sig.convolve2d(img, kernel_y, mode='same')
```

```python
# Plot!

fig = plt.figure(figsize=(18, 16), dpi= 80)
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(sobelG_x, cmap='gray')
ax2.imshow(sobelG_y, cmap='gray')
ax1.set_xlabel("sobelGx")
ax2.set_xlabel("sobelGy")
plt.show()
```
